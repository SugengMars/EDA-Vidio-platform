#### 1.4 Dataset Preprocessing


```python
import pandas as pd
from datetime import datetime

def extract_time(timestamp):
    timestamp = timestamp.rstrip('Z')  # Remove 'Z' character from the end of the string
    datetime_obj = datetime.fromisoformat(timestamp)
    time = datetime_obj.time()
    time_str = time.strftime('%H:%M:%S')
    return time_str
```

```python
df['Jam_mulai'] = df['play_time'].apply(extract_time)

# Menampilkan DataFrame hasil
df.head(2)
```

```python
df['Jam_selesai'] = df['end_time'].apply(extract_time)

# Menampilkan DataFrame hasil
df.head(2)
```

```python
def extract_month(timestamp):
    month = pd.to_datetime(timestamp.rstrip('Z')).month
    return month
```

```python
def extract_day(timestamp):
    month = pd.to_datetime(timestamp.rstrip('Z')).day
    return month
```

```python
df['Month'] = df['play_time'].apply(extract_month)
```

```python
df['Day'] = df['play_time'].apply(extract_day)
```
